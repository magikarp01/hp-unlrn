```python

import numpy as np
import json
import matplotlib.pyplot as plt

# Function to load and visualize npy file
def visualize_npy(npy_path):
    data = np.load(npy_path, allow_pickle=True)
    if data.ndim == 1:
        plt.plot(data)
        plt.title('1D Numpy Array')
    elif data.ndim == 2:
        plt.imshow(data, cmap='viridis', aspect='auto')
        plt.colorbar()
        plt.title('2D Numpy Array')
    else:
        # Add more conditions if you want to handle higher dimensional data differently
        print("Data has more than 2 dimensions and cannot be visualized directly.")
    plt.show()

# Function to load and summarize json file
def summarize_json(json_path):
    with open(json_path, 'r') as f:
        data = json.load(f)
        print(json.dumps(data, indent=2, sort_keys=True)[:500])  # Print first 500 characters of JSON string

# Function to display first few lines of a text file
def display_text_file(text_path):
    with open(text_path, 'r') as f:
        lines = f.readlines()
        print("".join(lines[:10]))  # Display first 10 lines

# Paths to the data files
npy_path = 'dicts_new.npy'  # Replace with your npy file path
json_paths = ['evaluation_prompts.json', 'probability_prompts.json']  # Replace with your json file paths
text_path = 'raw_data.txt'  # Replace with your text file path

# Visualize and summarize the data
# visualize_npy(npy_path)
for json_path in json_paths:
    summarize_json(json_path)
display_text_file(text_path)
```

```python
import nltk
from transformers import AutoTokenizer
import numpy as np

nltk.download('punkt')
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-chat-hf")

# Add padding token and resize token embeddings
# tokenizer.add_special_tokens({"pad_token": "<pad>"})
# pad_id = tokenizer.pad_token_id
pad_id = tokenizer.bos_token_id

def chunk_text(file_path, block_size=4096):
    # Read text file
    with open(file_path, 'r', encoding='utf-8') as f:
        text = f.read()

    # Split text into sentences
    sentences = nltk.tokenize.sent_tokenize(text)

    current_chunk = []
    current_length = 0
    for sentence in sentences:
        tokens = tokenizer.encode(sentence, add_special_tokens=False)
        token_length = len(tokens)

        if current_length + token_length > block_size:
            # Pad the current chunk if it's not the block size
            current_chunk.extend([pad_id] * (block_size - current_length))
            yield current_chunk
            current_chunk = tokens
            current_length = token_length
        else:
            current_chunk.extend(tokens)
            current_length += token_length

    # Pad and yield the last chunk if it's not empty
    if current_chunk:
        current_chunk.extend([pad_id] * (block_size - len(current_chunk)))
        yield current_chunk

# Example: process and save chunks
chunks = list(chunk_text("raw_data.txt"))

# Save chunks to npy file
np.save("rawdata_chunkedtokenized.npy", chunks)
```
